```python
sym =['*','+','!','@','^','$','/']
out = ""


for i in range(len(sym)):
    first = sym[i]
    for j in range(len(sym)):
        second = sym[j]
        for k in range(len(sym)):
            third = sym[k]
            out += first + second + third + "\n"
                           
print (out, end = "")

        
```
